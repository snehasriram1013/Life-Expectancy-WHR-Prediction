```python
# YOUR CODE HERE

X = df.drop(columns=['Healthy life expectancy at birth windsorized', 'Healthy life expectancy at birth'])
y = df['Healthy life expectancy at birth windsorized']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)

prediction = model.predict(X_test)
```

```python
#evaluation on the test set
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

```python
iterative_model = SGDRegressor(loss='squared_error', max_iter=1000, tol=1e-3, learning_rate='constant')

iterative_model.fit(X_train, y_train)

prediction_iter = iterative_model.predict(X_test)
```
